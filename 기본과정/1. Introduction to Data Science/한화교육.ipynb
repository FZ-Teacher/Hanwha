```python
kor_score = [39,69,20,100,80] 
math_score = [32,59,85,30,90] 
eng_score = [49,70,48,60,100] 
mid_score = [kor_score,math_score,eng_score] 

student_score=[0,0,0,0,0]
i=0
for subject in mid_score: # kor,math, eng :과목 선택
  for score in subject: #과목선택 후 점수를 선택.
    student_score[i] +=score # 각 학생마다 개별로 교과 점수를 저장
    print(subject,score,'i:',i,i,student_score)#kor->math->eng
    i+=1 #학생 index 구분
  i=0 # 과목이 바뀔 때 학생 인덱스
else: #for가 끝났을때 작동
  a,b,c,d,e = student_score #학생별 점수를 uppacking
  student_average = [a/3,b/3,c/3,d/3,e/3]
  print(student_average)
```

```python
def func1_with_enu(line):
  new_lines =[]
  for idx,val in enumerate(line): #index가 자동으로 부여가 됨.
    print('대기번호 %d번: %s' %(idx+1,val))
    new_lines.append((idx+1,val))
  return new_lines
```

```python
import numpy as np

def make_lotto(count):
  for i in range(count):
    lotto_num = [] #로또 번호가 담길 리스트 변수
    for j in range(6): #6번 반복
      lotto_num = np.random.choice(range(1,46),6,replace = False)
      lotto_num.sort()
    print('{}.로또 번호 : {}'.format(i+1,lotto_num))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi,np.pi,201)
y1 = np.sin(x)
y2 = np.cos(x)

fig,ax = plt.subplots()

ax.plot(x,y1,label ='사인함수')
ax.plot(x,y2,label ='코사인함수')

ax.legend(loc = 'lower right')

plt.show()
```
